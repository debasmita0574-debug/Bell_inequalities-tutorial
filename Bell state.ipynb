# Bell State: |Φ⁺⟩

In this notebook, we generate the Bell state:
(|00⟩ + |11⟩) / √2
This is the simplest example of quantum entanglement.

```python
!pip install qiskit qiskit-aer pylatexenc --quiet
```

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt
```

## Step 1: Create Bell State

We apply:
- Hadamard gate (H)
- CNOT gate (CX)

```python
qc = QuantumCircuit(2,2)

qc.h(0)
qc.cx(0,1)

qc.measure([0,1],[0,1])

qc.draw('mpl')
```

## Step 2: Run Simulation

```python
simulator = AerSimulator()
result = simulator.run(qc, shots=1024).result()
counts = result.get_counts()

print(counts)

plot_histogram(counts)
```

## Interpretation

We observe only:

00 and 11

This shows that:
- They are perfectly correlated

This is the signature of entanglement.
